```python
import pandas as pd
import numpy as np
results=pd.read_excel('Analytics_Attachment.xlsx', sheetname=1)
start_date = results['Date'][0]
results['Day Number']=((results['Date']-start_date)/np.timedelta64(1, 'D')).astype(int)
del results['Date']
results.loc[results['Winner'] == 'Home','Winning Team'] = results['Home Team']
results.loc[results['Winner'] == 'Home','Losing Team'] = results['Away Team']
results.loc[results['Winner'] == 'Home','Margin'] = (results['Home Score'] - results['Away Score']).astype(int)
results.loc[results['Winner'] == 'Away','Winning Team'] = results['Away Team']
results.loc[results['Winner'] == 'Away','Losing Team'] = results['Home Team']
results.loc[results['Winner'] == 'Away','Margin'] = (results['Away Score'] - results['Home Score']).astype(int)

results = results.merge(teams_divisions, left_on='Winning Team', right_on='Team_Name', how='left')
results=results.rename(columns = {'Division_id':'Winning Division','Conference_id':'Winning Conference'})
del results['Team_Name']
results = results.merge(teams_divisions, left_on='Losing Team', right_on='Team_Name', how='left')
results=results.rename(columns = {'Division_id':'Losing Division','Conference_id':'Losing Conference'})
del results['Team_Name']

results.loc[results['Winning Conference']==results['Losing Conference'], 'Conference'] = results['Winning Conference']
results.loc[results['Winning Conference']!=results['Losing Conference'], 'Conference'] = None
results.loc[results['Winning Division']==results['Losing Division'], 'Division'] = results['Winning Division']
results.loc[results['Winning Division']!=results['Losing Division'], 'Division'] = None
results=results.drop(['Winning Conference','Winning Division','Losing Conference','Losing Division', 'Home Team', 'Away Team',
                     'Home Score', 'Away Score', 'Winner'], axis=1)

results.set_index('Day Number')
```
